# Creating playlists and tracks dataframe

Function to extract tracks and audiofeatures

```python
def addtracks(tracks_json, tracklist, playlist):
    i=0
    for item in tracks_json['items']:
        if item['track']['id']:
            print('Extracting audiofeatures for track ', item['track']['id'], str(item['track']['name']))
            audio_features = sp.audio_features(item['track']['id'])
        else:
            audio_features = [None]
        if audio_features[0]:
            tracklist.append([playlist['id'], item['track']['id'], 
                              str(item['track']['name']), item['track']['artists'][0]['name'], 
                              item['track']['album']['release_date'], item['track']['popularity'],
                              audio_features[0]['danceability'], audio_features[0]['energy'],
                              audio_features[0]['key'],audio_features[0]['loudness'],
                              audio_features[0]['mode'],audio_features[0]['speechiness'],
                              audio_features[0]['acousticness'],audio_features[0]['instrumentalness'],
                              audio_features[0]['liveness'],audio_features[0]['valence'],
                              audio_features[0]['tempo'],audio_features[0]['duration_ms']])
        else:
            pass
        i+=1
```

Extracting tracks and playlist data for user 

```python
playlist_list = []
tracklist = []
for k in range(0,50):
    playlist = playlist_json['items'][k]
    print('Extracting playlist info', playlist['id'])
    if 'JGS' in playlist['name']:
        playlist_list.append([playlist['id'],str(playlist['name']),playlist['tracks']['total'],'JGS','jgs'])
    else:
        playlist_list.append([playlist['id'],str(playlist['name']),playlist['tracks']['total'],playlist['owner']['display_name'],playlist['owner']['id']])
    print('Adding tracks for playlist ', playlist['name'], playlist['id'])
    for i in range (0,int(int(playlist['tracks']['total'])/100)+1):
        j=0
        tracks_json = sp.playlist_tracks(playlist['id'], offset=i*100)
        addtracks(tracks_json, tracklist, playlist)
        print('-----------Extraction done-----------')


print('Creating playlist and tracks dataframes')
playlist_df=pd.DataFrame(data=playlist_list,columns=['playlist_id','playlist_name','ntracks','user_name','user_id'])
track_columns=['playlist_id', 'track_id','track_name', 
               'artist_name', 'release_date', 'popularity',
               'danceability','energy',
               'key','loudness',
               'mode','speechiness',
               'acousticness','instrumentalness',
               'liveness','valence',
               'tempo','duration_ms']

track_df=pd.DataFrame(data=tracklist,columns=track_columns)
print('-----------Dataframes created-----------')
```

To get the final dataset with all the tracks metrics and the playlist information we join the two datasets

```python
final_df = track_df.merge(playlist_df, on='playlist_id', how='left')
```
